### Exploring amount of tokens in articles

```python
import pandas as pd
```

```python
data_fakenewscorpus = pd.read_csv("../data/data_stemmed.csv", index_col=0)
data_fakenewscorpus["lengths"] = data_fakenewscorpus["content"].apply(lambda s: len(s.split(" ")))
```

```python
data_liar = pd.read_csv("../data/liar_stemmed_data.csv")
data_liar["lengths"] = data_liar["content"].apply(lambda s: len(s.split(" ")))
```

```python
# Average amount of tokens in an article
print((data_liar["lengths"].mean()))
print((data_fakenewscorpus["lengths"].mean()))
```
