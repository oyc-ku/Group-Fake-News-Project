```python
import pandas as pd
from fakenews_functions import clean_data, tokenize_and_remove_stopwords, stemming_words
```

```python
data=pd.read_csv('E:\KU_MachineLearning_Study\Fake_news_project\Group-Fake-News-Project\liar_test.tsv', sep='\t')
```

```python
old_header=data.columns
data=data.rename(columns={"true":"type",old_header[2]:"content"})
data.loc[len(data)]=old_header
```

```python
data["content"]=data["content"].apply(clean_data)
data["content"]=data["content"].apply(tokenize_and_remove_stopwords)
data["content"]=data["content"].apply(stemming_words)
```

```python
def change_label(label):
    if label in ["false","half-true","pants-fire","barely-true"]:
        return 0
    elif label in ["true","mostly-true"]:
        return 1
```

```python
data["type"]=data["type"].apply(change_label)
```

```python
data.to_csv("lair_data_cleaned.csv")
```
