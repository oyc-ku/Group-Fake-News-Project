```python
import pandas as pd
```

```python
datasetc=pd.read_csv("E:\KU_MachineLearning_Study\Fake_news_project\Group-Fake-News-Project\99950stemmedOvr.csv")
datasett=pd.read_csv("E:\KU_MachineLearning_Study\Fake_news_project\Group-Fake-News-Project\995,000_rows.csv")
dataset=datasetc
dataset["type"]=datasett["type"]
```

```python

class_list=["<email>","<url>","<num>","<date>"]
dict_info={}
for type in dataset["type"].unique():
	dict_info[type]={}
	for clas in class_list:
		dict_info[type][clas]=0
	dict_info[type]["amount"]=0
 
for index in range(len(dataset)):
    for clas in class_list:
        dict_info[dataset["type"].iloc[index]][clas]+=(dataset["content"].iloc[index]).count(clas)
    dict_info[dataset["type"].iloc[index]]["amount"]+=1
```

```python
for type in dataset["type"].unique():
    for clas in class_list:
        print(f"average amount of {clas} in {type} {(dict_info[type][clas])/(dict_info[type]["amount"])}")
```

```python
class_list=["<email>","<url>","<num>","<date>"]
dict_info_min={}
dict_info_max={}
for type in dataset["type"].unique():
    dict_info_max[type]={}
    dict_info_min[type]={}
    for clas in class_list:
        dict_info_min[type][clas]=999999999999999
        dict_info_max[type][clas]=-1
        
        dict_info_min[type][f"index{clas}"]=0
        dict_info_max[type][f"index{clas}"]=-1
        
for index in range(len(dataset)):
    for clas in class_list:
        if dict_info_min[dataset["type"].iloc[index]][clas]>dataset["content"].iloc[index].count(clas):
            dict_info_min[dataset["type"].iloc[index]][clas]=dataset["content"].iloc[index].count(clas)
            
        if dict_info_max[dataset["type"].iloc[index]][clas]<dataset["content"].iloc[index].count(clas):
            dict_info_max[dataset["type"].iloc[index]][clas]=dataset["content"].iloc[index].count(clas)
   
```

```python
for type in dataset["type"].unique():
    for clas in class_list:
        print(f"max amount of {clas} in {type} {dict_info_max[type][clas]}")
        print(f"min amount of {clas} in {type} {dict_info_min[type][clas]}")
        
```
