```python
import pandas as pd
from gensim.models import Word2Vec
```

```python
from tqdm import tqdm
tqdm.pandas()
```

```python
stemmed_data = pd.read_csv("./../data/data_stemmed.csv", index_col=0)
```

```python
class ArticleCorpus:
    def __init__(self, series):
        self.series = series

    def __iter__(self):
        for text in self.series:
            yield text.split()
```

```python
sentences = ArticleCorpus(stemmed_data["content"])
```

```python
model = Word2Vec(
    vector_size=128,
    window=5,
    min_count=5,
    workers=12,
    sg=1
)
```

```python
model.build_vocab(sentences, progress_per=10000)
```

```python
model.train(
    sentences,
    total_examples=model.corpus_count,
    epochs=5
)
```

```python
# TEST EMBEDDING MODEL
model.wv.most_similar("cheer", topn=10)
```

```python
model.save("../models/embedding_model.model")
```
