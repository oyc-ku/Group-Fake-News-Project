```python
import pandas as pd
```

```python
data=pd.read_csv("995,000_rows_stemming_words.csv")
```

```python
vectorizer = CountVectorizer()
```

```python
print(data.head())
```

```python
def change_label(label):
    if label in ["fake","clickbait","hate","rumor","unreliable","conspiracy","unknown"]:
        return "fake"
    else:
        return "reliable"
```

```python
data["type"]=data["type"].apply(change_label)
```

```python
X = vectorizer.fit_transform(data["content"].iloc[30:60])
y=data["type"].iloc[30:60]
#analyze = vectorizer.build_analyzer()
```

```python
print(y)
```

```python
model=clf = LogisticRegression(random_state=0).fit(X, y)
```
