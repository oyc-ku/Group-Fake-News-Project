```python
vectorizer = CountVectorizer(vocabulary=vocab_list)
```

```python
def change_label(label):
    if label in ["fake","clickbait","hate","rumor","unreliable","conspiracy","unknown"]:
        return "fake"
    else:
        return "reliable"
```

```python
data["type"]=data["type"].apply(change_label)
```

```python
train_set=data[0:int(len(data)*0.8)]
val_set=data[int(len(data)*0.8):int(len(data)*0.9)]
test_set=data[int(len(data)*0.9):]
```

```python
def remove_nan(x):
    if pd.isna(x):
        return " "
    else:
        return x
```

```python
train_set["content"]=train_set["content"].apply(remove_nan)
```
